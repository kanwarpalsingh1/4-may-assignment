Q1. Time Series and Its Applications:

A time series is a sequence of data points indexed in chronological order, typically collected at regular intervals over time.
Common applications of time series analysis include:
Financial forecasting (e.g., stock prices)
Demand forecasting (e.g., sales prediction)
Economic trend analysis
Weather forecasting
Health monitoring
Signal processing
Q2. Common Time Series Patterns:

Trend: Long-term increase or decrease in data values over time.
Seasonality: Regular patterns or cycles that repeat at fixed intervals.
Cyclical Patterns: Irregular patterns that occur over a span of multiple periods.
Randomness: Unpredictable fluctuations or noise in the data.
These patterns can be identified and interpreted using visual inspection, statistical tests, and decomposition techniques.
Q3. Time Series Data Preprocessing:

Handling missing values.
Removing outliers or anomalies.
Resampling or aggregation to adjust time intervals.
Differencing to stabilize non-stationary series.
Normalization or scaling to ensure uniformity across features.
Q4. Time Series Forecasting in Business Decision-Making:

Time series forecasting helps businesses anticipate future trends, make informed decisions, and plan resources effectively.
Challenges include handling seasonality, irregular patterns, and incorporating external factors such as economic conditions or market trends.
Limitations include the uncertainty of future events, model accuracy, and the assumption of stationary data.
Q5. ARIMA Modelling for Time Series Forecasting:

ARIMA (AutoRegressive Integrated Moving Average) models capture temporal dependencies in time series data.
They consist of autoregressive (AR), differencing (I), and moving average (MA) components to model trend, seasonality, and noise.
ARIMA models are widely used for forecasting stationary time series data.
Q6. ACF and PACF for Identifying ARIMA Orders:

ACF plots show autocorrelation at different lags, indicating the influence of past observations on current values.
PACF plots show partial autocorrelation, highlighting direct relationships between observations at different lags.
These plots help identify the order of ARIMA models by identifying significant lag values.
Q7. Assumptions of ARIMA Models:

Stationarity: Data should exhibit constant mean and variance over time.
Independence: Observations should be independent of each other.
Normality: Residuals should follow a normal distribution.
These assumptions can be tested using statistical tests and diagnostic plots.

Q9. Limitations of Time Series Analysis:

Time series analysis assumes that historical patterns will continue into the future, which may not always hold true.
It may struggle to account for sudden shifts or structural changes in data patterns, such as the impact of unforeseen events or market disruptions.
For example, during the COVID-19 pandemic, traditional time series models may have struggled to accurately forecast sales due to unprecedented changes in consumer behavior and economic conditions.
Q10. Stationary vs. Non-Stationary Time Series:
- Stationary Time Series: Mean, variance, and autocorrelation structure remain constant over time.
- Non-Stationary Time Series: Mean, variance, or autocorrelation change over time, indicating trends or seasonality.
- Stationarity affects the choice of forecasting model as many traditional methods, like ARIMA, assume stationarity to model temporal dependencies effectively. Non-stationary series often require preprocessing techniques like differencing to stabilize their properties before applying such models.
